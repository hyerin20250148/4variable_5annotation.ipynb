```python
numbers=[14,5,18,45,10,43]
for _ in range(len(numbers)):
  ret=numbers.pop()
  print(f'pop()으로추출된 값:{ret},numbers={numbers}')
```

```python
numbers=[14,5,18,45,10,43]
ret=numbers.pop(3)
print(f'pop()으로 추출된 값:(ret),numbers={numbers}')
```
