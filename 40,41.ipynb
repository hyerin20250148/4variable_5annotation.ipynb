```python
def MyRange(n):
  current=0
  while current<n:
    yield current
    current+=1

intList=MyRange(5)
print(list(intList))
```

```python
def myfunc(x):
  return x**2

results=map(myfunc,{0,1,2,3})
print(list(results))
```
