```python
class MyRange:
  def __init__(self, n):
    self.maxn = n
    self.current = 0

  def __iter__(self):
    return self

  def __next__(self):
    if self.current >= self.maxn:
      raise StopIteration
    else:
      ret = self.current
      self.current += 1
      return ret

intList = MyRange(5)
for i in intList:
  print(i, end='')
```

```python
def MyRange(n):
  current=0
  while current<n:
    yield current
    current+=1

  #제너레이터 객체 생성
  intList=MyRange(5)
  for i in intList:
    print(i,end='')
```
