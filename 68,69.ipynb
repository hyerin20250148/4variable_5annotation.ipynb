```python
mystack = []

def putData(item):
    mystack.append(item)
    print(f'스택에 데이터 추가: {item}')

def popData():
    if mystack:
        return mystack.pop()
    else:
        print('스택이 비어있습니다.')
        return None
```

```python
putData('사과')
putData([1,2,3])
putData(3.141592)
print(f'스택상태:{mystack}')
```

```python
ret=popData()
while ret!=None:
  print(f'스택에서 데이터 추출:{ret}')
  print(f'스택상태:{mystack}')
  print('-------------')
  ret=popData()
```
