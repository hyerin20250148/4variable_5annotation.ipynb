```python
def getPrime(x):
  if x ==2:
    return True
  if x<=1 or x%2==0:
    return False

  for i in range(3,int(x**(1/2))+1,2):
    if x%i==0:
      return False
    else:
      return True
```

```python
intList={x for x in range(1,101)}
ret=filter(getPrime,intList)
for p in ret:
  print(p,end='')
```
